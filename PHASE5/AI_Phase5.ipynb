```python
import pandas as pd

# Load the dataset (replace 'dataset.csv' with the actual file path)
data = pd.read_csv("C:\\Users\\Public\\Downloads\\COMED_hourly.csv\\COMED_hourly.csv")
```

```python
data['Datetime'] = pd.to_datetime(data['Datetime'])
data.set_index('Datetime', inplace=True)
print(data.columns)
# Calculate total energy consumption
total_energy_consumption = data['COMED_MW'].sum()
print(f"Total Energy Consumption: {total_energy_consumption} kWh")

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data.index, data['COMED_MW'], label='Energy Consumption (kWh)')
plt.title('Energy Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Energy Consumption (kWh)')
plt.legend()
plt.grid()
plt.show()
```
