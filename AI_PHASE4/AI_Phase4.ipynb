```python
import pandas as pd

# Load the dataset
data = pd.read_csv("C:\\Users\\prath\\Downloads\\PJME_hourly.csv")
```

```python
# Data exploration
print(data.info())
print(data.describe())
```

```python
# Convert 'Datetime' column to datetime
data['Datetime'] = pd.to_datetime(data['Datetime'])

# Handle missing values if needed
data = data.dropna()

# Resample data to daily frequency
daily_data = data.resample('D', on='Datetime').sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualization examples
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_data, x='Datetime', y='PJME_MW')
plt.title('Daily Energy Consumption')
plt.xlabel('Date')
plt.ylabel('Energy Consumption')
plt.grid(True)
plt.show()

# Create other visualizations (e.g., histograms, box plots, etc.)
```

```python
# Statistical analysis
mean_consumption = daily_data['PJME_MW'].mean()
median_consumption = daily_data['PJME_MW'].median()
std_deviation = daily_data['PJME_MW'].std()

print(f"Mean Energy Consumption: {mean_consumption}")
print(f"Median Energy Consumption: {median_consumption}")
print(f"Standard Deviation: {std_deviation}")

# Calculate correlations with other variables if applicable
```
